# 5.Analysis Questions

### a) Average Final Grade(G3)

```python
avg_grade=df['G3'].mean()
print('Average final grade:',round(avg_grade,2))
```

### b) Students Scoring Above 15

```python
above_15=len(df[df['G3']>15])
print('Students scoring above 15:',above_15)
```

### c) Correlation: Study Time vs Final Grade

```python
Correlation=df['studytime'].corr(df['G3'])
print('correlation between Study Time and Final Grade:',round(Correlation,3))
```

### d) Gender-Wise Average Perfomance

```python
gender_avg=df.groupby('sex')['G3'].mean()
gender_avg
```

# 6.Visualizations

### a) Histogram of Final Grades

```python
plt.figure()
plt.hist(df['G3'],bins=10)
plt.title('Distribution of Final Grades')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.show()
```

### b) Study Time vs Final Grade

```python
plt.figure()
plt.scatter(df['studytime'],df['G3'])
plt.title('Study Time vs Final Grade')
plt.xlabel('Study Time')
plt.ylabel('Final Grade(G3)')
plt.show()
```

### c) Male vs Female Average Score

```python
plt.figure()
gender_avg.plot(kind='bar')
plt.title('Average Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade')
plt.show()
```
